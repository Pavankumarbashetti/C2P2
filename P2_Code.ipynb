```python
import pandas as pd

url = "https://huggingface.co/datasets/theatticusproject/cuad/resolve/main/CUAD_v1/master_clauses.csv"
cuad_df = pd.read_csv(url)
print(cuad_df.columns, len(cuad_df))
cuad_df.head()
```

```python
# # 2. Example: filter only risky clauses (e.g., Indemnification, Termination)
# risky_cols = ["Indemnification", "Termination", "Change of Control"]
risky_cols = [
    "Indemnification Clause",                 # Not in your dataset but should be flagged if exists
    "Termination For Convenience",
    "Change Of Control",
    "Anti-Assignment",
    "Revenue/Profit Sharing",
    "Price Restrictions",
    "Minimum Commitment",
    "Volume Restriction",
    "Uncapped Liability",
    "Cap On Liability",
    "Liquidated Damages",
    "Irrevocable Or Perpetual License",
    "Source Code Escrow",
    "Post-Termination Services",
    "Audit Rights",
    "Non-Compete",
    "Exclusivity",
    "Most Favored Nation",
    "No-Solicit Of Customers",
    "No-Solicit Of Employees",
    "Non-Disparagement"
]

# Ensure we only select columns that exist in cuad_df
existing_cols = [col for col in risky_cols if col in cuad_df.columns]

print(existing_cols)

```

```python
# Check if filename column exists
# if "filename" in cuad_df.columns:
cols_to_select = ["Filename"] + existing_cols
# else:
#     cols_to_select = existing_cols
# Row count before filter
before_count = len(cuad_df)

# Apply filter
filtered_df = cuad_df[cols_to_select].dropna(how="all", subset=existing_cols)

# Row count after filter
after_count = len(filtered_df)

# Show comparison
print(f"Rows before filter: {before_count}")
print(f"Rows after filter:  {after_count}")
print(f"Rows removed:       {before_count - after_count}")

# Filter DataFrame
filtered_df = cuad_df[cols_to_select].dropna(how="all", subset=existing_cols)

filtered_df.head(5)

```

```python
# !pip install openpyxl
# Merge into text column for embeddings
filtered_df["text"] = filtered_df[existing_cols].apply(
    lambda row: " ".join(row.dropna().astype(str)), axis=1
)

len(filtered_df)
```
